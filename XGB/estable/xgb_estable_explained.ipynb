```python
import pandas as pd
import numpy as np
import xgboost as xgb
import shap

from sklearn.model_selection import GridSearchCV, StratifiedKFold

shap.initjs()
```

```python
# PATHS
DATA_PATH = r'../Data/pacientes_ucic_xgb.csv'

# VARIABLES
SCORING = 'accuracy'
CV = 5
RANDOM_STATE = 11
N_JOBS = -1
```

```python
# DATA
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
# X / y columns
y_col = 'Situación al alta de UCI.Estable no precisa cuidados especiales'
y_cols = [col for col in df.columns if 'Situación al alta' in col]
X_cols = df.drop(y_cols, axis=1).columns
```

```python
# X / y
X = df[X_cols]
y = df[y_col]
```

```python
PARAMS = {
    'max_depth': [2, 3],
    'learning_rate': [0.1, 0.01, 0.001],
    'n_estimators': [200, 300, 400],
    'scale_pos_weight': [y.value_counts()[1]/y.value_counts()[0], 1]
}
```

```python
skf = StratifiedKFold(n_splits=CV,
                      random_state=RANDOM_STATE)
```

```python
# BASE ESTIMATOR
clf = xgb.XGBClassifier(objective='binary:logistic', booster='gbtree', random_state=RANDOM_STATE, n_jobs=N_JOBS)
```

```python
grid_search = GridSearchCV(estimator=clf,
                          param_grid=PARAMS,
                          scoring=SCORING,
                          cv=CV,
                          n_jobs=N_JOBS)
grid_search.fit(X, y)
```

```python
best_params = grid_search.best_params_
best_params
```

```python
clf.set_params(**best_params)
clf.get_params()
```

```python
clf.fit(X, y)
```

```python
explainer = shap.TreeExplainer(clf)
```

```python
shap_values = explainer.shap_values(X)
```

```python
shap.summary_plot(shap_values, X)
```

```python
shap.summary_plot(shap_values, X, plot_type='bar')
```
